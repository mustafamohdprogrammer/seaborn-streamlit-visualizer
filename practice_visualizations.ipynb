```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the dataset
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
# Set seaborn theme
sns.set(style="whitegrid")
```

<h3> 1. Scatter Plot</h3>

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=tips, x="size", y="total_bill", hue="sex", markers=True)
plt.title("Scatter Plot - Total Bill vs Tip")
plt.show()
```

<h3> 2. Line Plot</h3>

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=tips, x="size", y="total_bill", hue="sex", markers=True)
plt.title("Line Plot - Size vs Total Bill by Sex")
plt.show()
```

<h3>3.Bar Plot</h3>

```python
plt.figure(figsize=(8,6))
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", palette="muted")
plt.title("Bar Plot - Total Bill by Day and Gender")
plt.show()
```

<h3>4. Box PLot</h3>

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=tips, x="day", y="tip", hue="smoker", palette="Set2")
plt.title("Box Plot - Tip by Day and Smoking Status")
plt.show()
```

<h3>5.Violin plot</h3>

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=tips, x="day", y="total_bill", hue="time", split=True, palette="pastel")
plt.title("Violin Plot - Total Bill by Day and Time")
plt.show()
```

<h3>6. Count plot</h3>

```python
plt.figure(figsize=(8,6))
sns.countplot(data=tips, x="day", hue="smoker", palette="dark")
plt.title("Count Plot - Days vs Smoker")
plt.show()
```

<h3>7. Regression plot</h3>

```python
plt.figure(figsize=(8,6))
sns.regplot(data=tips, x="total_bill", y="tip", scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title("Regression Plot - Total Bill vs Tip")
plt.show()
```

<h3>8. Histtogram</h3>

```python
plt.figure(figsize=(8,6))
sns.histplot(data=tips, x="total_bill", bins=20, kde=True, color='blue')
plt.title("Histogram of Total Bill with KDE")
plt.show()
```

<h3>9.Strip plotl</h3>

```python
plt.figure(figsize=(8,6))
sns.stripplot(data=tips, x="day", y="tip", hue="sex", jitter=True, palette="Set1")
plt.title("Strip Plot - Tips by Day and Gender")
plt.show()
```

<h3>10. Kde plot</h3>

```python
plt.figure(figsize=(8,6))
sns.kdeplot(data=tips, x="total_bill", hue="sex", fill=True, palette="tab10")
plt.title("KDE Plot - Total Bill Distribution by Gender")
plt.show()
```
